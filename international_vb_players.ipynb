```python
df = pd.read_csv('clean_data.csv')

df = df.sort_values(by='name').reset_index(drop=True)

player_info = df._get_numeric_data()


player_info 
```

```python
df.iloc[38]
```

```python
leyva_info = [player_info.iloc[38]]
print(leyva_info)
```

```python
from sklearn.metrics import pairwise_distances

similarity_score = pairwise_distances(leyva_info, player_info)
```

```python
player_name = df["name"].to_list()

similarity_list = similarity_score.tolist()


flat_similarity_list = []
for sublist in similarity_list :
  for item in sublist:
   flat_similarity_list.append(item)
flat_similarity_list
```

```python
player_name = df["name"].to_list()
```

```python
dictionary = dict(zip(player_name, flat_similarity_list))
print(dictionary)
```

```python
sorted_dict = dict( sorted(dictionary.items(),
                           key=lambda item: item[1],
                           reverse=False))

dict_items = sorted_dict.items()

five_most_similar = list(dict_items)[:10]
for key,value in five_most_similar:
  print(key , value)
```
